Задание 1**

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
def mark(rating):
    a, b, c = 'низкий рейтинг', 'средний рейтинг', 'высокий рейтинг'
    
    if rating <= 2:
        return a
    elif rating <= 4:
        return b
    else:
        return c

ratings['class'] = ratings['rating'].apply(mark)
ratings.head()
```

```python
new_table = ratings.merge(movie)
new_table = new_table.filter(items = ['rating', 'class' , 'title'])
new_table.head()
```

Задание 3 (бонусное)

В переменную years запишите список из всех годов с 1950 по 2010.

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

1. для каждой строки пройдите по всем годам списка years
2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
4. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
movie = pd.read_csv('C:\\Users\\Мария\\Desktop\\\ml-latest-small\\movies.csv',  encoding='utf-8')
ratings = pd.read_csv('C:\\Users\\Мария\\Desktop\\\ml-latest-small\\ratings.csv',  encoding='utf-8')
years = [x for x in range(1950 , 2011)]
 
movie.head()
```

```python
def production_year(title):
    lst = title.split(" ")
    year = lst[-1][1:-1]
    if year.isdigit(): 
        if int(year) in years:
            return year
    
    return 1900

movie['production_year'] = movie['title'].apply(production_year)
movie.tail(10)
```

```python
rate_year = movie.merge(ratings)
rate_year = rate_year.filter(items = ['rating', 'production_year'])
rate_year.groupby('production_year').mean().reset_index().sort_values('rating', ascending=False)


```
