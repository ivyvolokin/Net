```python
purchases = {}
with open('purchase_log.txt', encoding='utf-8') as f:
    f.readline()
    for line in f:
        
        line = line.strip('""').split()
         
        line = (line[1] + line[3]).strip('"}""').split(',')

        purchases[line[0]] = line[1]
```

```python
import json 

purchases = {}
with open('purchase_log.txt', encoding='utf-8') as f:
    
        f.readline()
        
        for line in f:
            dict_ = json.loads(line.strip())
            purchases[dict_['user_id']] = dict_['category']

# purchases
```
